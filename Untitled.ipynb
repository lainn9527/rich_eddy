```python
import pickle
import gzip
import talib
import numpy as np
import plotly.graph_objects as go
from strategy import strategy
from enum import IntEnum
```

```python
class TimePeriod(IntEnum):
    WEEK=5
    MONTH=20
    QUARTER=60
    YEAR=250
```

```python
stock_code = '2376'
with gzip.open('s_code_to_info.gz', 'rb') as fp:
    all_data: dict = pickle.load(fp)

data = all_data[stock_code]
```

```python
datetime = np.array([d[0] for d in data], dtype='datetime64')
close = np.array([d[4] for d in data], dtype=float)
ohlc = np.array([d[1:5] for d in data], dtype=float).transpose()
week_sma = talib.SMA(close, timeperiod = TimePeriod.WEEK)
month_sma = talib.SMA(close, timeperiod = TimePeriod.MONTH)
quarter_sma = talib.SMA(close, timeperiod = TimePeriod.QUARTER)
half_year_sma = talib.SMA(close, timeperiod = TimePeriod.QUARTER * 2)
year_sma = talib.SMA(close, timeperiod = TimePeriod.YEAR)
```

```python
fig = go.Figure(data=go.Ohlc(x=datetime, open=ohlc[0], high=ohlc[1], low=ohlc[2], close=ohlc[3]))
fig.show()
```
